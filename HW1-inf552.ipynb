```python
class1_data=biomed_data.iloc[:210]
class0_data=biomed_data.iloc[210:]
class_1_training=class1_data.iloc[0:140]
class_0_training=class0_data.iloc[0:70]
training_df=pd.concat([class_1_training,class_0_training])  #training dataset obtained by taking first 70 rows of class0
training_df.drop(['label'],axis='columns',inplace=True)      #and first 140 rows of class1

#training_df
class_1_testing=class1_data.iloc[140:]
class_0_testing=class0_data.iloc[70:]
testing_df=pd.concat([class_1_testing,class_0_testing])       #testing dataset which consists of the remaining rows
testing_df.drop(['label'],axis='columns',inplace=True)

```

```python

from sklearn.neighbors import KNeighborsClassifier  
classifier = KNeighborsClassifier(n_neighbors=3,metric='euclidean')  #classification based on Euclidean metric
classifier.fit(training_df.iloc[:,:-1],training_df['class'])  
```

```python
y_pred = classifier.predict(testing_df.iloc[:,:-1])  #prediction on testing dataset
print(y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix  #calculation of confusion matrix for k=3
print(confusion_matrix(testing_df['class'], y_pred)) 
print(classification_report(testing_df['class'], y_pred))
```

```python
import numpy as np
testerror=[]
list_of_k=[]
for i in range(208,0,-3):  #iterating over range of k in decrements of 3
    knn = KNeighborsClassifier(n_neighbors=i)    #classification for each k
    knn.fit(training_df.iloc[:,:-1],training_df['class'])
    pred_i = knn.predict(testing_df.iloc[:,:-1])
    testerror.append(np.mean(pred_i != testing_df['class'])) #obtaining the test error for every k
    list_of_k.append(i)                                       #adding all k values to a list
    testerror_kvalues=zip(list_of_k,testerror)
testerror_kvalues_list=list(testerror_kvalues)                #creating a list of (k,testerror) pairs
from operator import itemgetter        
print(sorted(testerror_kvalues_list,key=itemgetter(1)))       #sorting by the lowest test error
euc_result=min(testerror_kvalues_list,key=lambda x:x[1])      #storing the (k,test error) pair where test error is the lowest
print(euc_result)
```

```python
trainerror=[]
for i in range(208,0,-3):  
    knn = KNeighborsClassifier(n_neighbors=i)   
    knn.fit(training_df.iloc[:,:-1],training_df['class'])
    pred_i = knn.predict(training_df.iloc[:,:-1])
    trainerror.append(np.mean(pred_i != training_df['class'])) #creating a list of all the train errors
```

```python
from sklearn.neighbors import DistanceMetric

X_train=training_df.iloc[:,:-1]
y_train=training_df['class']
X_test=testing_df.iloc[:,:-1]
y_test=testing_df['class']
k_neighbors=list(range(1,200,5))
def classifier_create(neighbors,metric,X_train,y_train,p):  #function for creating the classifier
    classifier = KNeighborsClassifier(n_neighbors=neighbors,metric=metric,p=p)  
    classifier.fit(X_train,y_train) 
    return classifier
def predictor_test(classifier,X_test):     #function for predicting the class on test set
    y_predict = classifier.predict(X_test)
    return y_predict
def predictor_train(classifier,X_train):  #fucntion for predicting the class on train set
    y_p=classifier.predict(X_train)
    return y_p
def classification(y_test,y_train,neighbors,metric,p): #function for classification
    #testerrors=[]
    k_trainerror={}
    k_testerror={}
    for k in neighbors:
        knn = classifier_create(k,metric,X_train,y_train,p)
        pred_y_test= predictor_test(knn,X_test)
        pred_y_train=predictor_train(knn,X_train)
        k_trainerror[k]=np.mean(pred_y_train!=y_train)
        k_testerror[k]=np.mean(pred_y_test!= y_test)
    
        k_mintesterror=min(k_testerror.items(), key=lambda x: x[1])
        k_mintrainerror=min(k_trainerror.items(),key=lambda x: x[1])   
    
    return [k_mintesterror,k_mintrainerror]   #returns the list of (k,minimum test error) and (k,minimum train error)
        
```

```python
def log10p_classifier(y_test,neighbors,metric,p,p_errors): #function for classication for a range of log10p
    
   
    for k in neighbors:
        knn = classifier_create(k,metric,X_train,y_train,p)
        pred_y = predictor_test(knn,X_test)
        p_errors[p]=(np.mean(pred_y != y_test))
        p_minerror=min(p_errors.items(), key=lambda x: x[1]) #finding the best p and minimum testerror
        
    return(p_minerror)
```

```python
import math
log10p=list(np.arange(0.1,1.1,0.1))
bestp=[]
p_errors={}
for x in log10p:
    p=10**(x)  #finding p value
    p=round(p,1)
    #print(p)
    bestp_result=log10p_classifier(y_test,[6],'minkowski',p,p_errors)
log10v = round(math.log10(bestp_result[0]),1)  #finding the best log10p from best p
bestp.append(log10v)
bestp.append(bestp_result[1])

```

```python

def weighted_classifier_create(neighbors,metric,X_train,y_train,weights): #creating classifier for weighted decision polling
    classifier_w = KNeighborsClassifier(n_neighbors=neighbors,metric=metric,weights=weights)  
    classifier_w.fit(X_train,y_train) 
    return classifier_w
def w_predictor_test(classifier_w,X_test): #prediction for weighted decision polling(test data)
    y_predict_w = classifier_w.predict(X_test)
    return y_predict_w
def w_predictor_train(classifier_w,X_train): #prediction for weighted decision polling(train data)
    y_p_w=classifier_w.predict(X_train)
    return y_p_w
def w_classification(y_test,y_train,neighbors,metric,weights): #classification fuunction for different metrics and weighted 
    #testerrors=[]                                             #decision polling
    k_w_trainerror={}
    k_w_testerror={}
    for k in neighbors:
        knn_w = weighted_classifier_create(k,metric,X_train,y_train,weights)
        pred_y_test_w= w_predictor_test(knn_w,X_test)
        pred_y_train_w=w_predictor_train(knn_w,X_train)
        k_w_trainerror[k]=np.mean(pred_y_train_w!=y_train)
        k_w_testerror[k]=np.mean(pred_y_test_w!= y_test)
        k_mintesterror_w=min(k_w_testerror.items(), key=lambda x: x[1])
        k_mintrainerror_w=min(k_w_trainerror.items(),key=lambda x: x[1])   
        
    return [k_mintesterror_w,k_mintrainerror_w]
        
```
